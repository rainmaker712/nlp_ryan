````python
import re
#지금 추가로 하고싶은 건...?
#글자가 아닌 특수문자들을 없애고 싶습니다.
letters_only = re.sub("[^a-zA-Z]", " ", example1.get_text() )
#"[^a-zA-Z]": 대소문자 영어가 아닌(^) 친구들을 찾겠다는 의미
#" " : 대소문자가 아닌 친구들을 " " 공백으로 변경하고 싶다는 의미
#example1.get_text(): 이런 작업들을 수행하고 싶은 데이터
print ( letters_only )
````

````python
#매번 위와 같은 작업을 반복할 수 없으니 함수로 만들어봅니다.
def review_to_words( raw_review ):
    # 1. HTML 태그 지우기
    review_text = BeautifulSoup(raw_review, "html5lib").get_text()
    
    # 2. 영어가 아닌 특수문자들을 공백(" ")으로 바꾸기
    letters_only = re.sub("[^a-zA-Z]", " ", review_text)

    # 3. 대문자들을 소문자로 바꾸고 공백단위로 텍스트들을 나누기
    words = letters_only.lower().split()

    # 4. 파이썬에서 리스트보다 set에서의 찾기가 더 빠르다고 합니다.
    stops = set(stopwords.words("english"))
    
    # 5. 불용어가 아닌 것만 남기기
    meaningful_words = [w for w in words if not w in stops]
    
    # 6. 불용어가 아닌 텍스트를 공백 단위로 나누어서 리턴
    return( " ".join( meaningful_words ))
````

우리가 이 친구를 위해 하려는 작업은 다음과 같습니다.
다음과 같이 문장이 2개 있습니다.

문장1: "The cat sat on the hat"
문장2: "The dog ate the cat and the hat"

이 2개의 문장에서 우리는 단어를 전부 모아봅니다.

{ the, cat, sat, on, hat, dog, ate, and }

그리고 각 문장에서 위 단어 리스트들의 등장 횟수를 계산해봅니다.

문장1: { 2, 1, 1, 1, 1, 0, 0, 0 }
문장2: { 3, 1, 0, 0, 1, 1, 1, 1 }

우리가 위에서 열심히 작업했던 데이터는 예제 문장보다 훨씬 복작하기 때문에, 단어 리스트의 크기를 5000개의 자주 나오는 단어들을 선택해서 사용합니다.

````python
from sklearn.feature_extraction.text import CountVectorizer

print ("Creating the bag of words...\n")

#scikit-learn의 CountVectorizer 라는 친구가 우리가 하려는 작업을 도와줄겁니다.
vectorizer = CountVectorizer(analyzer = "word",
                             tokenizer = None,
                             preprocessor = None,
                             stop_words = None,
                             max_features = 5000)
````

````python
#위에 만든 모델에 우리의 데이터를 입력합니다. 우리의 단어 묶음들을 벡터화해줄겁니다.
train_data_features = vectorizer.fit_transform(clean_train_reviews)
````

````python
from sklearn.ensemble import RandomForestClassifier

#이번엔 Random Forest classifier(100 trees)를 사용해봅시다.
forest = RandomForestClassifier(n_estimators = 100) 

#단어 묶음을 벡터화한 데이터를 감정 label 데이터를 목표로하여 Fitting해봅시다.
forest = forest.fit( train_data_features, train["sentiment"] )
````

````python
#테스트 데이터를 읽어봅시다.
test = pd.read_csv("testData.tsv", header=0, delimiter="\t", quoting=3 )

#테스트 파일은 이렇게 생겼다고 합니다.
print (test.head())

#이 파일은 "sentiment" 행이 없습니다.
print (test.shape)

#불용어 제거도하고 태그들도 지우고 난후 데이터를 저장할 장소를 만들어둡니다.
num_reviews = len(test["review"])
clean_test_reviews = [] 

print ("Cleaning and parsing the test set movie reviews...\n")
for i in range(0,num_reviews):
    if( (i+1) % 1000 == 0 ):
        print ("Review %d of %d\n" % (i+1, num_reviews))
    clean_review = review_to_words( test["review"][i] )
    clean_test_reviews.append( clean_review )

#scikit-learn의 vectorizer를 활용해서 벡터화합니다.
test_data_features = vectorizer.transform(clean_test_reviews)
test_data_features = test_data_features.toarray()

#위에서 만들었던 Random Forest classifier란 친구를 활용해서 예측값을 뽑아봅니다.
result = forest.predict(test_data_features)

#알아보기 쉽게 데이터랑 붙여두는 편이 좋을 거 같습니다.
output = pd.DataFrame( data={"id":test["id"], "sentiment":result} )

#지금까지 처리한 결과를 파일로 저장합니다.
output.to_csv( "Bag_of_Words_model.csv", index=False, quoting=3 )
````
